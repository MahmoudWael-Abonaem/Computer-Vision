### Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read the Image

```python
Image = cv2.imread('...\Grayscale Image.jpg', cv2.IMREAD_GRAYSCALE)
```

### Display Original Image and its Histograms

```python
plt.title('Original Image')
plt.imshow(Image, cmap = 'gray')
plt.axis('off')

plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.hist(Image.flatten(), bins=256, range=(0, 255))
plt.title('Histogram of the Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2,2,2)
plt.hist(Image.flatten(), bins=256, range=(0, 255), cumulative=True, color='r')
plt.title('Cumulative Histogram of the Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Cumulative Frequency')

plt.show()
```

### Perform Intensity Shifting Operation

```python
Shifted_Image = np.clip(Image.astype(np.int32) + 60, 0, 255).astype(np.uint8)

plt.axis('off')
plt.title('Shifted Image')
plt.imshow(Shifted_Image, cmap = 'gray')

plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.hist(Shifted_Image.flatten(), bins=256, range=(0, 255))
plt.title('Histogram of the Shifted Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2,2,2)
plt.hist(Shifted_Image.flatten(), bins=256, range=(0, 255), cumulative=True, color='r')
plt.title('Cumulative Histogram of the Shifted Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Cumulative Frequency')

plt.show()
```

### Apply Histogram Equalization

```python
Equalized_Image = cv2.equalizeHist(Image)

plt.axis('off')
plt.title('Equalized Image')
plt.imshow(Equalized_Image, cmap = 'gray')

plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.hist(Equalized_Image.flatten(), bins=256, range=(0, 255))
plt.title('Histogram of Equalized Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2,2,2)
plt.hist(Equalized_Image.flatten(), bins=256, range=(0, 255), cumulative=True, color = 'r')
plt.title('Cumulative Histogram of Equalized Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Cumulative Frequency')

plt.show()
```

### Apply Contrast Stretching

```python
min_val = np.percentile(Image, 40)
max_val = np.percentile(Image, 100)

Stretched_Image = np.clip((Image.astype(np.float32) - min_val) * (255.0 / (max_val - min_val)), 0, 255).astype(np.uint8)

plt.imshow(Stretched_Image, cmap = 'gray')
plt.title('Stretched Image')
plt.axis('off')
plt.show()
```
