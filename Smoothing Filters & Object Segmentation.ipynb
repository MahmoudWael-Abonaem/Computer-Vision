### Imports

```python
import cv2
import numpy as np
from scipy.signal import wiener
from matplotlib import pyplot as plt
```

### Eight.tif Image

```python
eight_image_path = r'E:\Image Processing\Assign3\eight.tif'
original_eight_image = cv2.imread(eight_image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(6, 6))
plt.title('Original Image')
plt.imshow(original_eight_image, cmap='gray')
plt.axis('off')
plt.show()
```

### Apply Guassian Noise

```python
mean = 1
stddev = 1.5
gaussian_noise = np.random.normal(mean, stddev, original_eight_image.shape).astype(np.uint8)
noisy_image = cv2.add(original_eight_image, gaussian_noise)

# Display the noisy image
plt.figure(figsize=(5, 5))
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Image')
plt.axis('off')
plt.show()
```

### Apply Different Smoothing Filters (Average, Median and Mean)

```python
# Apply Average Filter
first_average_filter_size = (5, 5)
first_averaged_image = cv2.blur(noisy_image, first_average_filter_size)

second_average_filter_size = (3, 3)
second_averaged_image = cv2.blur(noisy_image, second_average_filter_size)

third_average_filter_size = (9, 9)
third_averaged_image = cv2.blur(noisy_image, third_average_filter_size)

# Display the image after applying the average filter
plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1)
plt.imshow(first_averaged_image, cmap='gray')
plt.title('After Applying 5×5 Average Filter')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(second_averaged_image, cmap='gray')
plt.title('After Applying 3×3 Average Filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(third_averaged_image, cmap='gray')
plt.title('After Applying 9×9 Average Filter')
plt.axis('off')

plt.show()
```

```python
# Apply Median Filter
first_median_filter_size = 5
first_medianed_image = cv2.medianBlur(noisy_image, first_median_filter_size)

second_median_filter_size = 3
second_medianed_image = cv2.medianBlur(noisy_image, second_median_filter_size)

third_median_filter_size = 9
third_medianed_image = cv2.medianBlur(noisy_image, third_median_filter_size)

# Display the image after applying the median filter
plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1)
plt.imshow(first_medianed_image, cmap='gray')
plt.title('After Applying 5×5 Median Filter')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(second_medianed_image, cmap='gray')
plt.title('After Applying 3×3 Median Filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(third_medianed_image, cmap='gray')
plt.title('After Applying 9×9 Median Filter')
plt.axis('off')

plt.show()
```

```python
# Apply Mean Filter
first_mean_size = 5
first_meaned_image = cv2.blur(noisy_image, (first_mean_size, first_mean_size))

second_mean_size = 3
second_meaned_image = cv2.blur(noisy_image, (second_mean_size, second_mean_size))

third_mean_size = 9
third_meaned_image = cv2.blur(noisy_image, (third_mean_size, third_mean_size))

# Display the image after applying the mean filter
plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1)
plt.imshow(first_meaned_image, cmap='gray')
plt.title('After Applying 5×5 Mean Filter')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(second_meaned_image, cmap='gray')
plt.title('After Applying 3×3 Mean Filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(third_meaned_image, cmap='gray')
plt.title('After Applying 9×9 Mean Filter')
plt.axis('off')

plt.show()
```

### Applying Adaptive Weiner Filter on the Eight.tif Image

```python
first_weiner_size = 1
first_wienered_image = wiener(noisy_image, (first_weiner_size, first_weiner_size))

second_weiner_size = 2
second_wienered_image = wiener(noisy_image, (second_weiner_size, second_weiner_size))

third_weiner_size = 3
third_wienered_image = wiener(noisy_image, (third_weiner_size, third_weiner_size))

# Display the image after applying the adaptive weiner filter
plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1)
plt.imshow(first_wienered_image, cmap='gray')
plt.title('After Applying 1×1 Adaptive Weiner Filter')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(second_wienered_image, cmap='gray')
plt.title('After Applying 2×2 Adaptive Weiner Filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(third_wienered_image, cmap='gray')
plt.title('After Applying 3×3 Adaptive Weiner Filter')
plt.axis('off')

plt.show()
```

### Circuit.tif Image

```python
circuit_image_path = r'E:\Image Processing\Assign3\circuit.tif'
original_circuit_image = cv2.imread(circuit_image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(5, 5))
plt.title('Original Image')
plt.imshow(original_circuit_image, cmap='gray')
plt.axis('off')
plt.show()
```

### Applying Sobel and Perwitt Filters on the Circuit.tif Image

```python
# Define The Perwitt Filter
perwitt_horizontal_filter = np.array([[-1, 0, 1],
                                     [-1, 0, 1],
                                     [-1, 0, 1]])

perwitt_vertical_filter = np.array([[-1, -1, -1],
                                     [0, 0, 0],
                                     [1, 1, 1]])

# Define The Sobel Filter
sobel_horizontal_filter = np.array([[-1, 0, 1],
                                    [-2, 0, 2],
                                    [-1, 0, 1]])

sobel_vertical_filter = np.array([[-1, -2, -1],
                                    [0, 0, 0],
                                    [1, 2, 1]])

# Apply Perwitt filter
perwitt_horizontal_filter1 = cv2.filter2D(original_circuit_image, -1, perwitt_horizontal_filter)
perwitt_vertical_filter1 = cv2.filter2D(original_circuit_image, -1, perwitt_vertical_filter)
perwitt_filtered_image = cv2.addWeighted(perwitt_horizontal_filter1, 0.5, perwitt_vertical_filter1, 0.5, 0)

# Apply Sobel filter
sobel_horizontal_filter1 = cv2.filter2D(original_circuit_image, -1, sobel_horizontal_filter)
sobel_vertical_filter1 = cv2.filter2D(original_circuit_image, -1, sobel_vertical_filter)
sobel_filtered_image = cv2.addWeighted(sobel_horizontal_filter1, 0.5, sobel_vertical_filter1, 0.5, 0)

# Plotting and comparing the results
plt.figure(figsize=(15, 10))

# Prewitt filtered
plt.subplot(1, 3, 2)
plt.imshow(perwitt_filtered_image, cmap='gray')
plt.title('After Applying Perwitt Filter')
plt.axis('off')

# Sobel filtered
plt.subplot(1, 3, 3)
plt.imshow(sobel_filtered_image, cmap='gray')
plt.title('After Applying Sobel Filter')
plt.axis('off')

plt.show()
```

### Apply Sobel Filter with Thresholding on the Circuit.tif Image

```python
# Apply Sobel filter
sobel_x = cv2.Sobel(original_circuit_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(original_circuit_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = np.sqrt(sobel_x ** 2 + sobel_y ** 2)


# Simple Thresholding
_, binary_threshold = cv2.threshold(sobel_magnitude, 50, 255, cv2.THRESH_BINARY)

# Convert sobel_magnitude to 8-bit unsigned integer
sobel_magnitude_uint8 = cv2.convertScaleAbs(sobel_magnitude)

# Apply adaptive thresholding
adaptive_threshold = cv2.adaptiveThreshold(sobel_magnitude_uint8, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 2)

plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(binary_threshold, cmap='gray')
plt.title('Binary Thresholding')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(adaptive_threshold, cmap='gray')
plt.title('Adaptive Thresholding')
plt.axis('off')

plt.show()
```

```python
# Read the grayscale Image
grayscale_image_path = r'E:\Image Processing\Grayscale Image.jpg'
Original_Grayscale_Image = cv2.imread(grayscale_image_path, cv2.IMREAD_GRAYSCALE)

# Apply Sobel filter
sobel_x = cv2.Sobel(Original_Grayscale_Image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(Original_Grayscale_Image, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = np.sqrt(sobel_x ** 2 + sobel_y ** 2)

# Simple Thresholding
_, binary_threshold = cv2.threshold(sobel_magnitude, 50, 255, cv2.THRESH_BINARY)

# Convert sobel_magnitude to 8-bit unsigned integer
sobel_magnitude_uint8 = cv2.convertScaleAbs(sobel_magnitude)

# Apply adaptive thresholding
adaptive_threshold = cv2.adaptiveThreshold(sobel_magnitude_uint8, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 2)

# Plotting and comparing the results
plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.imshow(Original_Grayscale_Image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_threshold, cmap='gray')
plt.title('Binary Thresholding')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(adaptive_threshold, cmap='gray')
plt.title('Adaptive Thresholding')
plt.axis('off')

plt.show()
```
