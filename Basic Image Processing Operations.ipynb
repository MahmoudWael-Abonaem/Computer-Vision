### Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read and Plot the Original Image

```python
image = cv2.imread('...\Grayscale Image.jpg')
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

### Applying Addition Operation

```python
# Define different constants to apply addition with different values
addition_constants = [50, 100, 150]

for constant in addition_constants:
    # Read the image
    image = cv2.imread('...\Grayscale Image.jpg')
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            new_pixel_value = image[i, j] + constant
            new_pixel_value = np.minimum(255, new_pixel_value)
            image[i, j] = new_pixel_value
            
    plt.title(f'Image with Addition Constant {constant}')
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

### Applying Subtraction Operation

```python
# Define different constants to apply subtraction with different values
subtraction_constants = [30, 60, 90]

for constant in subtraction_constants:
    # Read the image
    image = cv2.imread('...\Grayscale Image.jpg')
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            new_pixel_value = image[i, j] - constant
            new_pixel_value = np.maximum(0, new_pixel_value)
            image[i, j] = new_pixel_value

    plt.title(f'Image with Subtraction Constant {constant}')
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

### Applying Multiplication Operation

```python
# Define two different scaling factors (one is < 1 and the other is > 1) to apply them on the image
multiplication_factors = [0.5 , 1.5]

for factor in multiplication_factors:
    # Read the image
    image = cv2.imread('...\Grayscale Image.jpg')
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            new_pixel_value = image[i, j] * factor
            new_pixel_value = np.minimum(255, new_pixel_value)
            image[i, j] = new_pixel_value

    plt.title(f'Image with Subtraction Factor {factor}')
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

### Inverting the Image

```python
# Read the image
image = cv2.imread('...\Grayscale Image.jpg')

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        new_pixel_value = 255 - image[i, j]
        image[i, j] = new_pixel_value

plt.title('Image with Inverted Intensities')
plt.imshow(image)
plt.axis('off')
plt.show()
```
